```python
import numpy as np
x = np.linspace(0,5,11)
y =x**2
```

```python
x
```

```python
#functional method
```

```python
plt.plot(x,y, 'r-')
plt.xlabel('X label')
plt.title('Title')
plt.ylabel('Y label')
```

```python
#Multi plot
plt.subplot(1,2,1)
plt.plot(x,y,'r-')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
plt.xlabel('Y label')
plt.ylabel('X label')
```

```python
# Object oriented
fig = plt.figure() #blank canves
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) #left bottom width height
axes2 = fig.add_axes([0.2,0.3,0.4,0.3])
axes1.plot(x,y)
axes1.set_title('Larger plot')
axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
x = np.arange(0,100)
y = x*2
z= x**2
```

```python
#Exercise
```

```python
#1 OO
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title('title')
```

```python
#Functional
plt.plot(x,y)
plt.title('title')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])
ax1.plot(x,z)
ax1.set_xlabel('X')
ax1.set_ylabel('Z')

#insert
ax2.plot(x,y)
ax2.set_title('zoom')
ax2.set_xlim([20,22])
ax2.set_ylim([20,50])
```

```python
fig,axes=plt.subplots(1,2)
axes[0].plot(x,y,color='blue',lw=3,ls='--')
axes[1].plot(x,z,color='red',lw=3)
```
