```python
#Time Resampling
```

```python
df=pd.read_csv('walmart_stock.csv')
#df = pd.read_csv('walmart_stock.csv'),index_col = 'Date', parse_dates=True)
# if no parse, index_col will read'Date' as a string without transfering it to a datetime object
```

```python
df.head()
```

```python
df.info() #not a date time object but a string so we need to convert it
```

```python
df['Date']=pd.to_datetime(df['Date']) #date time index
```

```python
df['Date']=df['Date'].apply(pd.to_datetime) #Same as previous
```

```python
df.set_index('Date',inplace=True) #only for date time index
```

```python
df
```

```python
df.index
```

```python
df.resample(rule='Q').mean()
#Groupby method for year end frequency 'A'
#Quaterly resampling
```

```python
df.resample(rule='A').max()
```

```python
def first_day(entry):
    return entry[0]
```

```python
df.resample('A').apply(first_day)
#return very first entry for particular time period
```

```python
df['Close'].resample('M').mean().plot(kind='bar',figsize =(16,6))
```

```python
df.head()
```

```python
df.tail()
```

```python
df.shift(periods=1).head()
#lose tail end periods = 1, -1 lose first row
```

```python
df.head()
```

```python
df.shift(freq='M').head()
#Date time index gets shifted to last day of each Month, reassigning actual date time to match up with whatever frequency you provided
```

```python
#Rolling and expanding
```

```python
df.head()
```

```python
df['Open'].plot(figsize=(16,6))
```

```python
df.rolling(7).mean().head(20)
#Averaging previous 7 days, not a inplace regulation
#NaN cuz not enought data to calculate rolling mean
```

```python
df['Open'].plot()
df.rolling(window=30).mean()['Close'].plot(figsize=(16,6))
```

```python
df['Close 30 Day MA']= df['Close'].rolling(window=30).mean()
df[['Close 30 Day MA','Close']].plot(figsize=(16,6))
#Legend
```

```python
df['Close'].expanding().mean().plot(figsize=(16,6))
#general expanding mean from begining till end
```

```python
#Bollinger Band
#Volatility bands placed above and below a moving average where volatility is based on the standard deviation which changes as volatility increase or decreases and band will automatically widen when volatility increases and narrow when volatility decreases.
```

```python
#Close 20 MA
df['Close:20 Day Mean'] = df['Close'].rolling(20).mean()
#Upper = 20MA +2*std(20)
df['Upper'] = df['Close:20 Day Mean'] + 2*(df['Close'].rolling(20).std())
#Lower = 20MA - 2*std(20)
df['Lower'] = df['Close:20 Day Mean'] - 2*(df['Close'].rolling(20).std())
#Close
df[['Close','Close:20 Day Mean','Upper','Lower']].tail(200).plot(figsize=(16,6))
                                    
```
